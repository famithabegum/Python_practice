```python
class student:
    def __init__(self,name,roll_no,mark):
        self.name = name
        self.roll_no =roll_no
        self.mark = mark

students = [ student("Aslam",101,89),
           student("aysha",10,60),
           student("fami",103,90)]

for student in students:
    print(f" Name = {student.name}, Roll No: {student.roll_no}, Mark: {student.mark}")

```
