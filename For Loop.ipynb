```python
monthly_expenses=[2500,1800,5000,3000,4000]
Total_expense = 0
for i, expense in enumerate (monthly_expenses):
    print(f" month {i+1} ={expense}")
    Total_expense += expense
print("Total Expense =", Total_expense)
```
