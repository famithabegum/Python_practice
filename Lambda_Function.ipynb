```python
add = lambda a ,b : a+b

add(5,6)
```

```python
mul = lambda a,b,c: a*b*c 
mul (1,2,3)
```

```python
square = lambda a : a*a
square(5)
```

```python
even_odd = lambda a : "Even" if a%2 == 0 else "odd"
even_odd(9)
```

```python
numbers = [5,3,4,2]
result = list(map(lambda a : a*3 , numbers))
print(result)
```

```python
numbers = [2,3,5,6,7,8,11]
odd = list(filter(lambda a : a%2 !=0 , numbers))
print(odd)
```
